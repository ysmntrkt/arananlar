```python
Base = declarative_base()

class Terrorists(Base):
    __tablename__ = 'terrorists'
    id = Column(Integer, primary_key=True)
    name = Column(TEXT)
    surname = Column(TEXT)
    birthplace = Column(TEXT)
    birthday= Column(TEXT)
    organizationname=Column(TEXT)
    categoryid=Column(String)
    
    
    
    def __repr__(self):
        return "<terrorists(name='{}', surname='{}', birthplace='{}', birthday='{}',organizationname='{}',categoryid='{}')>"\
                .format(self.name, self.surname, self.birthplace, self.birthday,self.organizationname,self.categoryid)
Base.metadata.create_all(engine)
               
```

```python
class Photo(Base):
    __tablename__ = 'photos'
    id = Column(Integer, primary_key=True)
    image1=Column(String)
    image2=Column(String)
    image3=Column(String)
    terrorists_id = Column(Integer, ForeignKey('terrorists.id'),primary_key=True)
    terrorists=relationship("Terrorists", back_populates="photos", lazy="raise") 
    def __repr__(self):
        return "<photos(image1='{}',image2='{}',image3='{}',terrorists_id='{}')>"\
                .format(self.image)      

    
Terrorists.photos=  relationship("Photo", order_by=Photo.id, back_populates="terrorists")               
Base.metadata.create_all(engine)                
```
