```python
engine = create_engine('postgresql+psycopg2://postgres:1111@localhost:5432/arananlar_db',encoding='utf-8-sig', echo=True,executemany_mode='values_plus_batch',executemany_values_page_size=10000, executemany_batch_page_size=500 )
Session = sessionmaker(engine)
s = Session()
```

```python
Base = declarative_base()

class Terrorists(Base):
    __tablename__ = 'terrorists'
    id = Column(Integer, primary_key=True)
    name = Column(TEXT)
    surname = Column(TEXT)
    birthplace = Column(TEXT)
    birthday= Column(String)
    organizationname=Column(TEXT)
    categoryid=Column(String)
s.autoflush = True    
    #photo=relationship("Photo", backref="terrorists", lazy=True)
    
def __repr__(self):
        return "<terrorists(name='{}', surname='{}', birthplace='{}', birthday='{}',organizationname='{}',categoryid='{})>"\
                .format(self.name, self.surname, self.birthplace, self.birthday,self.organizationname,self.categoryid)
Base.metadata.create_all(engine)
```
